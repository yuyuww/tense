```python

#%% 0.1. 대용량 데이터를 90~100만 개 단위로 CSV 파일로 저장하는 함수
def save_dataframe_to_csv(df, docu_column, dir_path, file_prefix, batch_size=900000, batch_size2=1000000):
    """
    대용량 데이터프레임을 일정 크기로 나누어 CSV 파일로 저장하는 함수.

    Parameters:
    - df: 저장할 데이터프레임
    - docu_column: 문서 기준 컬럼 (ex: 'docu_id'), 같은 문서는 같은 파일에 저장
    - dir_path: 저장할 폴더 경로
    - file_prefix: 저장 파일명 접두사
    - batch_size: 최소 행 수 (900,000개) (문서가 끊기지 않으면 더 저장)
    - batch_size2: 최대 행 수 (1,000,000개) (최대 크기 초과 시 강제 분할)

    Returns:
    - docu_df: 문서 ID와 저장된 파일명을 매핑한 데이터프레임
    """

    # 저장할 폴더 설정
    folder_name = "word_csv"
    save_dir = os.path.join(dir_path, folder_name)
    os.makedirs(save_dir, exist_ok=True)  # 폴더가 없으면 생성

    # 예외 처리: 데이터프레임이 비어 있을 경우
    if df.empty:
        print("⚠️ 저장할 데이터가 없습니다.")
        return pd.DataFrame()

    # 파일 저장 함수 (중복 코드 제거를 위해 내부 함수화)
    def _save_file(start_idx, end_idx, file_number):
        file_name = f"{file_prefix}_{file_number}.csv" #파일명 만들기
        file_path = os.path.join(save_dir, file_name) #파일 경로 생성
        df.iloc[start_idx:end_idx].to_csv(file_path, index=False) #csv파일 저장
        print(f"   ✅ Saved {file_name}")

        # docu_column이 어느 파일에 저장되었는지 기록
        docu = df.iloc[start_idx:end_idx][docu_column].drop_duplicates().values
        return pd.DataFrame({docu_column: docu, 'csv_file_name': file_name}) #결과값으로 docu_column + 파일명 df 반환

    # 초기 변수 설정
    file_name_number, start_idx = 1, 0
    docu_list = []

    while start_idx < len(df):
        next_idx = min(start_idx + batch_size, len(df))  # 기본적인 batch_size 단위로 분할
        current_docu = df.iloc[next_idx - 1][docu_column]  # 문서 ID 확인

        # 같은 문서가 끝날 때까지 확장 (최대 batch_size2 초과하지 않도록)
        while next_idx < len(df) and df.iloc[next_idx][docu_column] == current_docu:
            next_idx += 1
            if (next_idx - start_idx) > batch_size2:
                break

        # 파일 저장 및 문서 목록 업데이트
        docu_list.append(_save_file(start_idx, next_idx, file_name_number))

        start_idx = next_idx
        file_name_number += 1

    # 모든 파일에 대한 문서 저장 정보 반환
    return pd.concat(docu_list, ignore_index=True)
```

```python
#sen_id열 변환
# sen_id 변환 함수: ["AA0001.1", "AA0001.2", "BB1002.12", "CC2023.5", "DD9999.123"] -> ["AA0001.00001", "AA0001.00002", "BB1002.00012", "CC2023.00005", "DD9999.00123"]
def convert_sen_id(sen_id):
    # '.'을 기준으로 앞(문자열)과 뒤(번호)를 분리
    base, num = sen_id.rsplit(".", 1)  # 오른쪽에서 한 번만 split (마지막 '.' 처리)
    
    # 뒤쪽 숫자를 5자리 정수형 문자열로 변환
    num_formatted = f"{int(num):05d}"  # 1 → 00001, 12 → 00012, 123 → 00123
    
    # 새로운 sen_id 생성
    return f"{base}.{num_formatted}"

# 데이터프레임에 적용
if "sen_id" in df.columns:
    df["sen_id"] = df["sen_id"].apply(convert_sen_id)
```

```python
#ID열 변환
if "word_id2" in df.columns and "sen_id" in df.columns:
    # word_id2를 6자리 문자열로 변환 (앞에 0을 채움)
    df["word_id2"] = df["word_id2"].astype(str).str.zfill(4)

    # 새로운 ID 열 생성 (sen_id.word_id2)
    df["ID"] = df["sen_id"].astype(str) + "." + df["word_id2"]
df["word_id2"] = df["word_id2"].astype(int)
```
